```python
#1:file opener 

#open playlist txt file: store the file handler in a variable
fh = open("top100_2019.txt")
#initialize playlist name
playlist = "Ben's 100 songs of 2019"
#read through playlist.txt lines 
lines = fh.readlines()

#create function to parse different playlist txt files 
def parse_playlist(playlist_entry):
    #save as elements, lowercase all characters, remove whitespace, split the sline into two elements at the"-"
    elements = playlist_entry.lower().strip().split("-")
    #initalize dictionary
    artist_to_songs = {
        #first element is the artist 
        "artist": elements[0].strip(),
        #second element is the song titles
        "songs": elements[1]
    }
    #create function to call as artist_to_songs
    return artist_to_songs
```

```python
#2: dictionary
    
# define an index to map counts to addresses
artist_songs = {}


#3: filter
    
# for each line in the file
for line in lines:
    #use the function with parsed elements
    element = parse_playlist(line)
        
    #get the artist name as song_list and define list for song list     
    songs_list = artist_songs.get(element["artist"], [])
    #add songs to list if is for certain artist 
    songs_list.append(element['songs'])
    #update the dictionary for each artist and their specific songs
    artist_songs.update({element['artist']: songs_list})


#4: counter 
    
#start count at 0 
count = 0 
#loop through songs 
for x in artist_songs:
    #find conditions 
    if isinstance (artist_songs[x], list):
        #add length of list in list 
        count += len(artist_songs[x])
        

#5: printer

#print statment for total songs 
print ("In '" + playlist + "' there are " + str(count) + " songs.")
#skip line for easier reading 
print ()
#print statement for each artist 
for artist in artist_songs:
    print ( artist  +  "'s songs are:" + str(artist_songs[artist]))
    print ()
```

```python
#6: song counter for artists in playlist 
#print statement for playlist title and introduction
print ("Song count for each artist in '"+ playlist + "':")
#skip line for easier reading 
print()
#loop through artists list of songs
for k,v in artist_songs.items():
    #print each artist with count of each item in list of artist 
    print (k, len(list(filter(None, v))))
```

```python
#search query 

#user input, store as query variable 
query = input("Enter artist name to find their song(s) in '" + playlist + "': ")
#lowercase query 
query = query.lower()

#results key and value 
    
#get artist songs based on query, show list of songs 
results = artist_songs.get(query, [])

#printer
    
#print statement with length of songs and artist name again 
print ("There is " + str(len(results)) + " of " + str(query) + "'s song(s) in this playlist:")

#print each song title 
for result in results:
    #go through each song and print it 
    print (result)
```
